```python
def eval_DCRBase(dataset_name):
    #Load the dataset
    x, c, y = get_dataset(dataset_name)

    #Load the model
    cem, dcr = load_DCRBase(x, c, y, dataset_name)

    #Create the dataframe
    df, classes = create_df_DCRBase(x, cem, dcr)

    #Plot the concept importance
    #plot_feature_importance(df, dataset_name)

    #Plot the concept importance for each class
    plot_feature_importance_per_class(df, classes, dataset_name)
```
